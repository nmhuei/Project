```python
# ====== 1. DATA ======
X = [
    [30, 3, 6],
    [43, 4, 8],
    [25, 2, 3],
    [51, 4, 9],
    [40, 3, 5],
    [20, 1, 2]
]

y = [2.5, 3.4, 1.8, 4.5, 3.2, 1.6]

# ====== 2. SCALE (optional nhưng nên có) ======
X = [[x/50, f/5, r/10] for x,f,r in X]

# ====== 3. MATRIX ======
X_mat = matrix(RDF, X)
y_vec = vector(RDF, y)

# ====== 4. ADD BIAS ======
m = X_mat.nrows()
X_bias = matrix(RDF, [[1] + list(row) for row in X_mat])

# ====== 5. INIT THETA ======
theta = zero_vector(RDF, 4)

# ====== 6. COST FUNCTION ======
def compute_cost(X, y, theta):
    m = len(y)
    pred = X * theta
    return sum((pred[i] - y[i])^2 for i in range(m)) / (2*m)

# ====== 7. GRADIENT DESCENT ======
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    
    for i in range(iterations):
        pred = X * theta
        error = pred - y
        
        grad = (X.transpose() * error) / m
        theta = theta - alpha * grad
        
        # debug nhẹ cho đỡ mù
        if i % 1000 == 0:
            print(f"Iter {i}, cost = {compute_cost(X, y, theta)}")
    
    return theta

# ====== 8. TRAIN ======
theta = gradient_descent(X_bias, y_vec, theta, alpha=0.01, iterations=10000)

print("\nTheta:", theta)

# ====== 9. PREDICT ======
def predict(size, floors, rooms):
    # scale giống training
    x = vector(RDF, [1, size/50, floors/5, rooms/10])
    return x * theta

print("\nTest predict:", predict(35, 3, 6))
```
